#### Q4. 
* Crie um dataframe `gols` com duas colunas `pa√≠s` e `total_gols` com o total de gols marcados por cada pa√≠s durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
pa√≠s        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python

# Gols time 1
gols_time1 = wc.groupby('time_1')['gols_1'].sum()
# Gols time 2
gols_time2 = wc.groupby('time_2')['gols_2'].sum()

# Combinar
gols_total = pd.DataFrame({
    'pa√≠s': list(set(gols_time1.index) | set(gols_time2.index))
})
gols_total['total_gols'] = gols_total['pa√≠s'].apply(
    lambda x: gols_time1.get(x, 0) + gols_time2.get(x, 0)
)

# Ordenar
gols = gols_total.sort_values('total_gols', ascending=False).reset_index(drop=True)

print("\nü•Ö Top 10 pa√≠ses com mais gols:")
print(gols.head(10).to_string(index=False))
````

#### Q5. 
Qual pa√≠s tomou mais cart√µes amarelos somando todas as copas?

Neste exerc√≠cio voc√™ vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observa√ß√£o √© uma string represetando uma lista dos cart√µes amarelos de um √∫nico jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16‚Äô|1:0|Rosana G√≥mez', '20‚Äô|2:0|Gabriela Ch√°vez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o m√©todo gen√©rico `apply` para chamar uma fun√ß√£o que voc√™ vai criar para transformar uma observa√ß√£o de cart√£o amarelo em uma contagem de elementos da lista. 

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Em seguida fa√ßa sua m√°gica para agrupar as informa√ß√µes por pa√≠s, acumular os cart√µes de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
pa√≠s        cart√µes amarelos
Argentina   97
England     93
Australia   93
...
```

````python
# Fun√ß√£o para contar cart√µes
def contar_cartoes(valor):
    if pd.isna(valor):
        return 0
    # Removendo os colchetes e conta as v√≠rgulas + 1
    lista_str = str(valor).strip("[]'")
    if lista_str == '' or lista_str == 'nan':
        return 0
    return len([x for x in lista_str.split("', '") if x.strip()])

# Aplicar fun√ß√£o
wc['num_cartoes_1'] = wc['cartao_amarelo_1'].apply(contar_cartoes)
wc['num_cartoes_2'] = wc['cartao_amarelo_2'].apply(contar_cartoes)

# Somar cart√µes por pa√≠s
cartoes_time1 = wc.groupby('time_1')['num_cartoes_1'].sum()
cartoes_time2 = wc.groupby('time_2')['num_cartoes_2'].sum()

# Combinar
cartoes_total = pd.DataFrame({
    'pa√≠s': list(set(cartoes_time1.index) | set(cartoes_time2.index))
})
cartoes_total['cart√µes_amarelos'] = cartoes_total['pa√≠s'].apply(
    lambda x: cartoes_time1.get(x, 0) + cartoes_time2.get(x, 0)
)

# Ordenar
cartoes = cartoes_total.sort_values('cart√µes_amarelos', ascending=False).reset_index(drop=True)

print("\nüü® Top 10 pa√≠ses com mais cart√µes amarelos:")
print(cartoes.head(10).to_string(index=False))
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de p√™nalti.

Para conseguir essa informa√ß√£o, voc√™ precisar√° trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas tamb√©m s√£o colunas textuais, onde cada observa√ß√µes apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol est√° na forma `'Jogador(a) ¬∑ minuto‚Äô'`. Por exemplo:
```
'Alex Morgan ¬∑ 12‚Äô|Rose Lavelle ¬∑ 20‚Äô'
```

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Recomendo criar um dicion√°rio √† parte, onde cada chave ser√° um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente ser√° a contagem de ocorr√™ncias desses nomes.

Em seguida basta converter o seu dicion√°rio em um dataframe e imprim√≠-lo na forma (valores n√£o s√£o inventados üòÅ):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
# Dicion√°rio para armazenar gols
jogadores_gols = {}

# Processar colunas de gols
colunas_gols = ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']

for col in colunas_gols:
    for valor in wc[col]:
        if pd.isna(valor) or str(valor) == 'nan':
            continue
        
        # Processar string
        gols_lista = str(valor).split('|')
        for gol in gols_lista:
            if ' ¬∑ ' in gol or ' √Ç¬∑ ' in gol:
                # Extrair nome do jogador
                separador = ' ¬∑ ' if ' ¬∑ ' in gol else ' √Ç¬∑ '
                jogador = gol.split(separador)[0].strip().strip("[]'")
                
                if jogador and jogador != '':
                    jogadores_gols[jogador] = jogadores_gols.get(jogador, 0) + 1

# Criar DataFrame
gols_jogadores = pd.DataFrame(list(jogadores_gols.items()), 
                              columns=['jogador(a)', 'num_gols'])
gols_jogadores = gols_jogadores.sort_values('num_gols', ascending=False).reset_index(drop=True)

print("\n‚öΩ Top 10 jogadores com mais gols:")
print(gols_jogadores.head(10).to_string(index=False))
````
